```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

Učitajmo podatke u padnas objekt.

```python
data = pd.read_csv("../results.txt")
data.head(10)
```

Pogledajmo najprije razliku između rezultata kod reduce i bez reduce parametra.

Najzanimljivije usporedbe bi tu bile između `per_fill` i:
+ `num_of_insertion_call`
+ `sum_MNK`
+ `insertion_time`
+ `row_fill`

```python
sb.set_theme()
fill = data.loc[data.length_of_sequence == 10, :].sort_values(by=['per_fill'])

f, ax = plt.subplots(2, 2, figsize=(10, 10))

sb.lineplot(data=fill, x='per_fill', y='num_of_insertion_call', hue='reduce',
            style='reduce',
            ax=ax[0, 0],
            markers=True).set(title='K=10, Percentage fill and number of insertion')

sb.lineplot(data=fill, x='per_fill', y='sum_MNK', hue='reduce',
            style='reduce',
            ax=ax[0, 1],
            markers=True).set(title='K=10, Percentage fill and MNK')

sb.lineplot(data=fill, x='per_fill', y='insertion_time', hue='reduce',
            style='reduce',
            ax=ax[1, 0],
            markers=True).set(title='K=10, Percentage fill and insertion time')

sb.lineplot(data=fill, x='per_fill', y='row_fill', hue='reduce',
            style='reduce',
            ax=ax[1, 1],
            markers=True).set(title='K=10, Percentage fill and rows that are filled')

# plt.savefig('Figure_01.png')
# plt.savefig('Figure_01.svg')
```

Pogledajmo sada zanimljiva područja koja se nalaze od 0.7 do 1.0.

```python
sb.set_theme()
fill = data.loc[data.length_of_sequence == 10, :].sort_values(by=['per_fill'])

f, ax = plt.subplots(2, 2, figsize=(10, 10))

sb.lineplot(data=fill, x='per_fill', y='num_of_insertion_call', hue='reduce',
            style='reduce',
            ax=ax[0, 0],
            markers=True).set(title='K=10, Percentage fill and number of insertion')
ax[0, 0].set_xlim(0.8, 1)

sb.lineplot(data=fill, x='per_fill', y='sum_MNK', hue='reduce',
            style='reduce',
            ax=ax[0, 1],
            markers=True).set(title='K=10, Percentage fill and MNK')
ax[0, 1].set_xlim(0.8, 1)

sb.lineplot(data=fill, x='per_fill', y='insertion_time', hue='reduce',
            style='reduce',
            ax=ax[1, 0],
            markers=True).set(title='K=10, Percentage fill and insertion time')

sb.lineplot(data=fill, x='per_fill', y='row_fill', hue='reduce',
            style='reduce',
            ax=ax[1, 1],
            markers=True).set(title='K=10, Percentage fill and rows that are filled')
ax[1, 1].set_xlim(0.8, 1)

# plt.savefig('Figure_02.png')
# plt.savefig('Figure_02.svg')
```

Pogledajmo sada parametar `K` kako utječe na parove popunjenosti i:

+ `insertion_time`

```python
fill = data.loc[data.reduce == 0, :]
fill['size_'] = fill.loc[:, 'num_of_buckets'] * fill.loc[:, 'bucket_size']
fill = fill.sort_values(by=['size_'])

plt.figure(figsize=(10, 5))
sb.lineplot(data=fill, x='size_', y='row_fill', hue='length_of_sequence',
            style='length_of_sequence',
            markers=True).set(title='K=10, Percentage fill and number of insertion')
plt.xlim(5 * 1e+6, 6 * 1e+6)

# plt.savefig('Figure_03.png')
# plt.savefig('Figure_03.svg')

plt.figure(figsize=(10, 5))
sb.lineplot(data=fill, x='size_', y='sum_MNK', hue='length_of_sequence',
            style='length_of_sequence',
            markers=True).set(title='K=10, Percentage fill and number of insertion')
plt.xlim(5 * 1e+6, 5.5 * 1e+6)

# plt.savefig('Figure_04.png')
# plt.savefig('Figure_04.svg')

plt.figure(figsize=(10, 5))
sb.lineplot(data=fill, x='size_', y='num_of_insertion_call', hue='length_of_sequence',
            style='length_of_sequence',
            markers=True).set(title='K=10, Percentage fill and number of insertion')
plt.xlim(5 * 1e+6, 6 * 1e+6)

# plt.savefig('Figure_05.png')
# plt.savefig('Figure_05.svg')
```
