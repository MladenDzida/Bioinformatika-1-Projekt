```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv("../results.txt")
data.head(10)
```

```python
fill = data.loc[data.length_of_sequence == 10]
fill['size'] = data.loc[:, 'num_of_buckets'] * data.loc[:, 'bucket_size']
fill.head(10)
```

```python
fill = fill.sort_values(by=['size'])
fill.head(10)
```

```python
plt.plot(fill.loc[:, 'size'], fill.loc[:, 'per_fill'])
plt.xlabel('Size')
plt.ylabel('Fill')
plt.savefig('10_size_per_fill.png')
plt.savefig('10_size_per_fill.svg')
```

```python
sb.set_theme()
sb.lineplot(data=fill, x='size', y='per_fill')
```

```python
def plot(array_1, array_2, data):
    data['size'] = data.loc[:, 'num_of_buckets'] * data.loc[:, 'bucket_size']
    data = data.sort_values(by=['size'])
    data['per_row_fill'] = data.loc[:, 'row_fill'] / data.loc[:, 'num_of_buckets']
    
    for x, y in zip(array_1, array_2):
        plt.figure();
        sb.lineplot(data=data, x=x, y=y, hue='length_of_sequence', style='reduce', markers=True)

array_1 = ['size', 'size']
array_2 = ['per_fill', 'per_row_fill']
plot(array_1, array_2, data.loc[(data.length_of_sequence == 100)])
```
